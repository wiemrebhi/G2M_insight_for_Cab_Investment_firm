# **Machine Learning to Predict Taxi Fare**





```python
data=df1[["Date of Travel","City","KM Travelled","Price Charged"]]
data
```

```python
data["Date of Travel_year"] = data["Date of Travel"].dt.year
data["Date of Travel_month"] = data["Date of Travel"].dt.month
data["Date of Travel_week"] = data["Date of Travel"].dt.day
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['City']=le.fit_transform(data['City'])
```

```python
from matplotlib import pyplot

data.corr(method='spearman').style.format("{:.2}").background_gradient(cmap=pyplot.get_cmap('coolwarm'))

```

```python
data
```

```python
from sklearn.model_selection import train_test_split
X=data.drop(["Price Charged","Date of Travel"],axis=1)
y=data["Price Charged"]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20,random_state=123)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_train, y_train)
```

```python
y_pred=reg.predict(X_test)
y_pred
```

```python
import math
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

Random forest


```python
from sklearn.ensemble import RandomForestClassifier
parameters = {'n_estimators':[200,300,400,500], 
              'max_features': ['sqrt','log2'],
               'max_depth' : [3,4,5,6,7],
               'random_state' : [18]}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor


grid_search = GridSearchCV(estimator = RandomForestRegressor(), 
                           param_grid = parameters, 
                           cv = 5)
```

```python
grid_search.fit(X_train,y_train)
grid_search.best_params_
```

```python
rf = grid_search.best_estimator_
```

```python
grid_search.best_score_
```

```python
y_pred = rf.predict(X_test)
y_pred
```

```python
import math
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

 Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

depth = list(range(3,30))
param_grid = dict(max_depth = depth)
tree = GridSearchCV(DecisionTreeRegressor(), param_grid, cv = 10)
tree.fit(X_train,y_train)
tree.best_params_
```

```python
tree.best_score_
```

```python
tr= tree.best_estimator_
```

```python
y_pred = tr.predict(X_test)
y_pred
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

Ridge Regression


```python
from sklearn.linear_model import Ridge
```

```python
params = {'alpha' : [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]}
ridge_regressor = GridSearchCV(Ridge(), params, cv = 5, scoring = 'neg_mean_absolute_error', n_jobs = -1)
ridge_regressor.fit(X_train, y_train)
```

```python
ridge_regressor.best_params_

```

```python
ridge_regressor.best_score_
```

```python
rr= ridge_regressor.best_estimator_
```

```python
y_pred = rr.predict(X_test)
y_pred
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

Lasso Regression

```python
from sklearn.linear_model import Lasso
```

```python
# Performing GridSearchCV on Lasso Regression
params = {'alpha' : [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]}
lasso_regressor = GridSearchCV(Lasso(), params ,cv = 15,scoring = 'neg_mean_absolute_error', n_jobs = -1)
lasso_regressor.fit(X_train, y_train)
```

```python
lasso_regressor.best_params_
```

```python
lasso_regressor.best_score_
```

```python
lr= lasso_regressor.best_estimator_
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
MSE = mean_squared_error(y_test, y_pred)
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

Comparing Models

```python
ridge_score = round(rr.score(X_train, y_train) * 100, 2)
ridge_score_test = round(rr.score(X_test, y_test) * 100, 2)

lasso_score = round(lr.score(X_train, y_train) * 100, 2)
lasso_score_test = round(lr.score(X_test, y_test) * 100, 2)

decision_score = round(tr.score(X_train, y_train) * 100, 2)
decision_score_test = round(tr.score(X_test, y_test) * 100, 2)

random_forest_score=round(rf.score(X_train, y_train) * 100, 2)
random_forest_score_test=round(rf.score(X_test, y_test) * 100, 2)

linear_regression_score=round(reg.score(X_train, y_train) * 100, 2)
linear_regression_score_test=round(reg.score(X_test, y_test) * 100, 2)
```

```python
models = pd.DataFrame({
    'Model': [ 'Ridge Regression', 'Lasso Regression','Decision Tree Regressor','Random Forest Regression','Linear regression'],
    'Score': [ ridge_score, lasso_score, decision_score,random_forest_score,linear_regression_score],
    'Test Score': [ ridge_score_test, lasso_score_test, decision_score_test,random_forest_score_test,linear_regression_score_test]})
models.sort_values(by='Test Score', ascending=False)
```
